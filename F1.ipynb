```python
all_laps = []
all_weather = []

gp_name = 'Italian Grand Prix'

for year in range(2018, 2025):
    print(f"\n--- Processando ano: {year} ---")

    sessions_to_load_this_year = {}
    if year == 2021:
        sessions_to_load_this_year = {
            'Race': sessions_monza_race,
            'Sprint': sessions_monza_sprint,
            'Qualifying': sessions_monza_qualifying,
            'FP1': sessions_monza_fp1,
            'FP2': sessions_monza_fp2,
        }
    else:
        sessions_to_load_this_year = {
            'Race': sessions_monza_race,
            'Qualifying': sessions_monza_qualifying,
            'FP1': sessions_monza_fp1,
            'FP2': sessions_monza_fp2,
            'FP3': sessions_monza_fp3,
        }

    for session_name, session_list in sessions_to_load_this_year.items():
        try:
            session = fastf1.get_session(year, gp_name, session_name)
            session.load()
            session_list.append(session)

            laps = session.laps.copy()
            laps['Year'] = year
            all_laps.append(laps)

            weather = session.weather_data
            if not weather.empty:
                weather['Year'] = year
                all_weather.append(weather)

            print(f"  -> Sucesso: '{session_name}' de {year} carregado e dados extraídos.")


        except Exception as e:
            print(f"  -> Falha: Não foi possível carregar '{session_name}' de {year}. (Erro: {e})")

if all_laps:
    df_laps = pd.concat(all_laps, ignore_index=True)
    df_weather = pd.concat(all_weather, ignore_index=True)


    print(f"Total de {len(df_laps)} voltas de corrida coletadas.")
    print(f"Total de {len(df_weather)} registros meteorológicos coletados.")

    df_laps = df_laps.sort_values(by='LapStartTime').reset_index(drop=True)
    df_weather = df_weather.sort_values(by='Time').reset_index(drop=True)


    df_merged = pd.merge_asof(
        left=df_laps,
        right=df_weather,
        left_on='LapStartTime',
        right_on='Time',
        by='Year',
        direction='nearest'
    )


    df_merged['FuelLoad'] = 110 - (df_merged['LapNumber'] * 1.6)


    df_merged = pd.get_dummies(df_merged, columns=['Compound', 'Driver'], prefix=['Compound', 'Driver'])

    df_merged.dropna(subset=['LapTime'], inplace=True)
    df_clean = df_merged[df_merged['IsAccurate'] == True].copy()


    df_clean['LapTimeSeconds'] = df_clean['LapTime'].dt.total_seconds()

    df_clean.dropna(subset=['LapTimeSeconds'], inplace=True)
    df_clean = df_clean[df_clean['LapTimeSeconds'] < 200]

    features_to_exclude = [
      'Time_x',
      'Time_y',
      'LapStartTime',
      'LapStartDate',
      'Sector1SessionTime',
      'Sector2SessionTime',
      'Sector3SessionTime',
      'FastF1Generated',
      'IsPersonalBest',
      'FreshTyre',
      'Deleted',
      'DeletedReason',
      'TrackStatus',
      'Team',
      'Compound_nan',
      'Compound_UNKNOWN',
      'PitOutTime',
      'PitInTime',
      'Position',
      'Pressure',
      'Humidity',
      'WindDirection',
      'WindSpeed',
      'DriverNumber'
    ]

    df_final = df_clean.drop(columns=features_to_exclude, errors='ignore')

    if sessions_monza_race:
        first_race = None
        for session in sessions_monza_race:
            if not session.results.empty:
                first_race = session
                break

        if first_race:
            winner = first_race.results.iloc[0]['FullName']
            print(f"O vencedor da corrida de {first_race.event['EventDate'].year} foi {winner}.")
        else:
            print("Nenhuma sessão de corrida carregada tinha dados de resultados.")


else:
    print("Nenhum dado de corrida foi carregado para processar.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
```

```python
cut_off_year = 2024
train_df = df_final[df_final['Year'] < cut_off_year].copy()
test_df = df_final[df_final['Year'] == cut_off_year].copy()

print(len(train_df))
print(len(test_df))

target = 'LapTimeSeconds'

print(train_df.columns)

model_features = [col for col in train_df.columns if col != 'Year']


X_train_df = train_df[model_features]
y_train_df = train_df[target]

X_test_df = test_df[model_features]
y_test_df = test_df[target]
```

```python
def create_sequences(X_data, y_data, sequence_length=10):
    X_sequences, y_sequences = [], []
    for i in range(len(X_data) - sequence_length):
        X_sequences.append(X_data[i:(i + sequence_length)])
        y_sequences.append(y_data[i + sequence_length])

    return np.array(X_sequences), np.array(y_sequences)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train_df)

target_col_index = X_train_df.columns.get_loc(target)


all_X_train_seq, all_y_train_seq = [], []

for year in train_df['Year'].unique():
    year_df = train_df[train_df['Year'] == year]

    X_year_scaled = scaler.transform(year_df[model_features])

    y_year_scaled = X_year_scaled[:, target_col_index]

    X_seq, y_seq = create_sequences(X_year_scaled, y_year_scaled, sequence_length=10)

    all_X_train_seq.append(X_seq)
    all_y_train_seq.append(y_seq)

X_train = np.concatenate(all_X_train_seq)
y_train = np.concatenate(all_y_train_seq)

X_test_scaled = scaler.transform(X_test_df)
y_test_scaled = X_test_scaled[:, target_col_index]
X_test, y_test = create_sequences(X_test_scaled, y_test_scaled, sequence_length=10)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
sequence_length = X_train.shape[1]
n_features = X_train.shape[2]

model = Sequential(name="Modelo_F1_LSTM")

model.add(LSTM(units=128, activation='tanh', return_sequences=True, input_shape=(sequence_length, n_features)))
model.add(Dropout(0.2))

model.add(LSTM(units=64, activation='tanh'))
model.add(Dropout(0.2))

model.add(Dense(units=1, activation='linear'))

optimizer = tf.keras.optimizers.Adam(learning_rate=0.0005)
model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mean_absolute_error'])

model.summary()
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001)
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, shuffle=False, callbacks=[early_stopping, reduce_lr])
```

```python
scaled_predictions = model.predict(X_test)
feature_columns = X_train_df.columns
dummy_array = np.zeros((len(scaled_predictions), len(feature_columns)))
target_col_index = feature_columns.get_loc('LapTimeSeconds')
dummy_array[:, target_col_index] = scaled_predictions.flatten()

full_unscaled_data = scaler.inverse_transform(dummy_array)

final_predictions_in_seconds = full_unscaled_data[:, target_col_index]

print(final_predictions_in_seconds[:5])
```

```python
y = test_df[target].values[sequence_length:]
mae = mean_absolute_error(y, final_predictions_in_seconds)
mse = mean_squared_error(y, final_predictions_in_seconds)
rmse = np.sqrt(mse)
print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
```

```python
feature_order = X_train_df.columns.tolist()
target_col_index = feature_order.index('LapTimeSeconds')

TOTAL_LAPS = 53
PIT_STOP_TIME_LOSS = 23.0
DRIVER_TO_SIMULATE = 'VER'
RACE_CONDITIONS = {
    'TrackTemp': 42.0,
    'AirTemp': 28.0,
    'Humidity': 50.0,
    'Pressure': 1012.0,
    'WindSpeed': 6.0,
    'Rainfall': 0.0
}
```

```python
def simulate_stint(start_lap, end_lap, initial_tyre_life, compound_name, driver_name,
                   initial_fuel_load, model, scaler, feature_order, race_conditions):

    predicted_lap_times = []

    history_df = X_test_df.iloc[0:10].copy()

    current_lap = start_lap

    for i in range(end_lap - start_lap + 1):
        history_scaled = scaler.transform(history_df[feature_order])
        input_sequence = np.expand_dims(history_scaled, axis=0)

        scaled_prediction = model.predict(input_sequence, verbose=0)[0][0]

        dummy_array = np.zeros((1, len(feature_order)))
        dummy_array[0, target_col_index] = scaled_prediction
        unscaled_prediction_array = scaler.inverse_transform(dummy_array)
        predicted_time = unscaled_prediction_array[0, target_col_index]
        predicted_lap_times.append(predicted_time)

        next_lap_data = {}
        for col in feature_order:
            if 'Compound_' in col:
                next_lap_data[col] = 1 if col == f'Compound_{compound_name}' else 0
            elif 'Driver_' in col:
                next_lap_data[col] = 1 if col == f'Driver_{driver_name}' else 0
            elif col in race_conditions:
                next_lap_data[col] = race_conditions[col]
            else:
                next_lap_data[col] = history_df[col].iloc[-1]

        next_lap_data['LapNumber'] = current_lap
        next_lap_data['TyreLife'] = initial_tyre_life + i
        next_lap_data['LapTimeSeconds'] = predicted_time
        next_lap_data['FuelLoad'] = initial_fuel_load - (current_lap * 1.6)

        next_lap_df = pd.DataFrame([next_lap_data], columns=feature_order)

        history_df = pd.concat([history_df.iloc[1:], next_lap_df], ignore_index=True)

        current_lap += 1

    return predicted_lap_times
```

```python
one_stop_results = {}
two_stop_results = {}

strategy_1_stop = ['HARD', 'MEDIUM']
for pit_lap in range(20, 36):
    stint1 = simulate_stint(1, pit_lap, 0, strategy_1_stop[0], DRIVER_TO_SIMULATE, 110.0, model, scaler, feature_order, RACE_CONDITIONS)
    stint2 = simulate_stint(pit_lap + 1, TOTAL_LAPS, 0, strategy_1_stop[1], DRIVER_TO_SIMULATE, 110.0, model, scaler, feature_order, RACE_CONDITIONS)
    total_time = sum(stint1) + PIT_STOP_TIME_LOSS + sum(stint2)
    one_stop_results[pit_lap] = total_time

strategy_2_stops = ['MEDIUM', 'MEDIUM', 'SOFT']
for pit1_lap in range(15, 26):
    for pit2_lap in range(pit1_lap + 15, 42):
        stint1 = simulate_stint(1, pit1_lap, 0, strategy_2_stops[0], DRIVER_TO_SIMULATE, 110.0, model, scaler, feature_order, RACE_CONDITIONS)
        stint2 = simulate_stint(pit1_lap + 1, pit2_lap, 0, strategy_2_stops[1], DRIVER_TO_SIMULATE, 110.0, model, scaler, feature_order, RACE_CONDITIONS)
        stint3 = simulate_stint(pit2_lap + 1, TOTAL_LAPS, 0, strategy_2_stops[2], DRIVER_TO_SIMULATE, 110.0, model, scaler, feature_order, RACE_CONDITIONS)
        total_time = sum(stint1) + sum(stint2) + sum(stint3) + (2 * PIT_STOP_TIME_LOSS)
        two_stop_results[(pit1_lap, pit2_lap)] = total_time
```

```python
best_1_stop_lap = min(one_stop_results, key=one_stop_results.get)
best_1_stop_time = one_stop_results[best_1_stop_lap]

best_2_stops_laps = min(two_stop_results, key=two_stop_results.get)
best_2_stops_time = two_stop_results[best_2_stops_laps]

print(f"Melhor Estratégia de 1 Parada (D->M): Parar na volta {best_1_stop_lap}. Tempo total: {best_1_stop_time/60:.2f} min")
print(f"Melhor Estratégia de 2 Paradas (M->M->M): Parar nas voltas {best_2_stops_laps[0]} e {best_2_stops_laps[1]}. Tempo total: {best_2_stops_time/60:.2f} min")

if best_1_stop_time < best_2_stops_time:
    print(f"A ESTRATÉGIA ÓTIMA GERAL é de 1 PARADA, parando na volta {best_1_stop_lap}.")
else:
    print(f"A ESTRATÉGIA ÓTIMA GERAL é de 2 PARADAS, parando nas voltas {best_2_stops_laps[0]} e {best_2_stops_laps[1]}.")
```
